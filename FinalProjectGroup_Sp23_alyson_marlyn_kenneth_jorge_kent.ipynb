# Data Cleaning

The final dataset was created by merging the datasets, mentioned in the *Datasets* section, with additional features. A description of this process, along with the necessary code, is detailed below.

### Load Datasets

```python
# Import raw datasets 

gdp = pd.read_csv('gdp.csv')
gdp_per_cap = pd.read_csv('gdp_per_cap.csv')
gdp_growth = pd.read_csv('gdp_growth.csv')
gdp_growth_per_cap = pd.read_csv('gdp_growth_per_cap.csv')
co2_emissions = pd.read_csv('co2_emissions.csv')
life_expec = pd.read_csv('life_expectancy.csv')
net_migration = pd.read_csv('net_migration.csv')
unemployment = pd.read_csv('unemployment.csv')
foreign_aid = pd.read_csv('us_foreign_aid_country.csv')
```

### Removing unnecessary columns

```python
# Remove unnecessary columns, make sure you only run once or you will keep removing 3 columns 
# Dates should start at 1960

gdp = gdp.drop(gdp.columns[1:4], axis = 1)
gdp_per_cap = gdp_per_cap.drop(gdp_per_cap.columns[1:4], axis = 1)
gdp_growth = gdp_growth.drop(gdp_growth.columns[1:4], axis = 1)
gdp_growth_per_cap = gdp_growth_per_cap.drop(gdp_growth_per_cap.columns[1:4], axis = 1)
co2_emissions = co2_emissions.drop(co2_emissions.columns[1:4], axis = 1)
life_expec = life_expec.drop(life_expec.columns[1:4], axis = 1)
net_migration = net_migration.drop(net_migration.columns[1:4], axis = 1)
unemployment = unemployment.drop(unemployment.columns[1:4], axis = 1)

foreign_aid = foreign_aid.drop(['Country Code', 'Region ID', 'Region Name', 'Income Group Acronym', 
                  'Income Group Name', 'Transaction Type ID'], axis = 1)
```

### Extracting the 10 nations of interest and the U.S.

```python
# Filter data to only contain the 10 nations of interest and the U.S. for comparison
# Rename countries so they match the other datasets 

foreign_aid['Country Name'] = foreign_aid['Country Name'].replace(['Syria', 'Egypt', 'Congo (Kinshasa)'], ['Syrian Arab Republic', 'Egypt, Arab Rep.', 'Congo, Dem. Rep.'])

top_10 = ['Afghanistan', 'Israel', 'Jordan', 'Egypt, Arab Rep.', 'Ethiopia', 'Iraq', 'Nigeria', 'South Africa', 
          'Congo, Dem. Rep.', 'Syrian Arab Republic', 'United States'] # list of countries of interest

gdp = gdp[gdp['Country Name'].isin(top_10)]
gdp_per_cap = gdp_per_cap[gdp_per_cap['Country Name'].isin(top_10)]
gdp_growth = gdp_growth[gdp_growth['Country Name'].isin(top_10)]
gdp_growth_per_cap = gdp_growth_per_cap[gdp_growth_per_cap['Country Name'].isin(top_10)]
co2_emissions = co2_emissions[co2_emissions['Country Name'].isin(top_10)]
life_expec = life_expec[life_expec['Country Name'].isin(top_10)]
net_migration = net_migration[net_migration['Country Name'].isin(top_10)]
unemployment = unemployment[unemployment['Country Name'].isin(top_10)]

foreign_aid = foreign_aid[foreign_aid['Country Name'].isin(top_10)]
```

### Changing the data format 

```python
# Change the format from wide to long, combines the dates into a single column and the indicators

gdp = gdp.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'GDP')
gdp_per_cap = gdp_per_cap.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'GDP_Per_Cap')
gdp_growth = gdp_growth.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'GDP_Growth')
gdp_growth_per_cap = gdp_growth_per_cap.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'GDP_Growth_Per_Cap')
co2_emissions = co2_emissions.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'Co2_Emissions')
life_expec = life_expec.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'Life_Expec')
net_migration = net_migration.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'Net_Migration')
unemployment = unemployment.melt(id_vars = 'Country Name', var_name = 'Year', value_name = 'Unemployment')
```

### Merging data on Year and Country name

```python
# Create list of data frames to merge them all at once. Merge on Year and Country Name

dfs = [gdp, gdp_per_cap, gdp_growth, gdp_growth_per_cap, co2_emissions, life_expec, net_migration, unemployment]
data = reduce(lambda  left,right: pd.merge(left,right,on=['Year', 'Country Name'],
                                            how='outer'), dfs)
data
```

### Cleaning foreign_aid dataset 

```python
# Clean foreign_aid before merging with data 

# Only necessary transaction type, represents the total amount that was actually given to the nation by the U.S. 
foreign_aid = foreign_aid[foreign_aid['Transaction Type Name'] == 'Disbursements']

# Drop columns. 'constant_amount' accounts for inflation, this will align with the rest of the data
foreign_aid = foreign_aid.drop(['Transaction Type Name', 'current_amount'], axis = 1)

# Rename columns to match rest of data 
foreign_aid = foreign_aid.rename(columns = {'Fiscal Year' : 'Year', 'constant_amount' : 'Total_Aid'})
```

### Merge data and foreign_aid

```python
# Merge the data
complete_data = pd.merge(data, foreign_aid, on = ['Year', 'Country Name'])
complete_data
```

### Create separate dataframe that contains just the U.S. 

```python
# Separate df that contains only the U.S. 

usa = data[(data['Country Name'] == 'United States') & (data['Year'].astype(int) >= 2001)]
usa.head()
```

### Save data as csv 

```python
# Save as a csv
complete_data.to_csv('clean_data.csv', index = False)
usa.to_csv('usa.csv',index = False)
```

# Multivariate Regression

```python
# Log all of the variables before running linear models

columns = ['GDP', 'GDP_Per_Cap', 'GDP_Growth', 'GDP_Growth_Per_Cap',
       'Co2_Emissions', 'Life_Expec', 'Net_Migration', 'Unemployment',
       'Total_Aid']

for i in columns:
    clean_data[i + '_log'] = np.log10(clean_data[i])
clean_data
```

### Scatterplot of total aid against each indicator

```python
# Scatterplot of total aid against each indicator

outcomes = ['GDP_log', 'GDP_Per_Cap_log', 'GDP_Growth_log', 'GDP_Growth_Per_Cap_log',
              'Co2_Emissions_log', 'Life_Expec_log', 'Net_Migration_log', 'Unemployment_log']

for outcome in outcomes:
    sns.scatterplot(x='Total_Aid_log', y=outcome, data=clean_data)
    plt.xlabel('Total Aid (log)')
    plt.ylabel(outcome)
    plt.title(f'Total Aid vs. {outcome}')
    plt.show()
```

Plotting the data points on the above scatterplots, we can visualize the distribution of the data by metric measured against the total foreign aid.

### What are the overall R-squared values across all countries and all metrics? 

```python
# Ols model to determine relationship between indicators and total aid
# Plot of R-Squared values for each indicator 

outcomes = ['GDP_log', 'GDP_Per_Cap_log', 'GDP_Growth_log', 'GDP_Growth_Per_Cap_log',
              'Co2_Emissions_log', 'Life_Expec_log', 'Net_Migration_log', 'Unemployment_log']

r_squared = []

for outcome in outcomes:
    formula = outcome + '~ Total_Aid_log'
    model = smf.ols(formula, data = clean_data).fit()
    r_square = model.rsquared
    r_squared.append(r_square)
    
# Plot of R-Squared values 
plt.figure(figsize=(10,5))
plt.bar(outcomes, r_squared)
plt.xlabel('Indicators')
plt.ylabel('R-Squared')
plt.title('R-Squared Values for Each Indicator')
plt.xticks(rotation = 70)
plt.show()
```

Based on the bar chart above, the highest R-squared correlation value is observed with unemployment (R-square value of ~0.023) and life expectancy (R-square value of ~0.027). However, overall across all countries, with the highest R-square value being ~0.027, the data suggests that there is no strong correlation across the countries between the amount of foreign aid received and the metrics that we are measuring against. Now, let's take a deeper look into the countries themselves to see if we can conclude anything else.

```python
# Linear model of total aid against each indicator by each country if R-Squared is above 0.5

ideal_r = 0.7 or -0.7 # Only plotting variables that have a strong positive or negative relationship 
                     # (too many to plot at once)

outcomes = ['GDP_log', 'GDP_Per_Cap_log', 'GDP_Growth_log', 'GDP_Growth_Per_Cap_log',
              'Co2_Emissions_log', 'Life_Expec_log', 'Net_Migration_log', 'Unemployment_log']

for outcome in outcomes:
    for country in clean_data['Country Name'].unique():
        country_data = clean_data[clean_data['Country Name'] == country]
        filtered_data = country_data.dropna(subset=[outcome])
        
        formula = outcome + '~ Total_Aid_log'
        model = smf.ols(formula, data = filtered_data).fit()
        r_squared = model.rsquared 
        
        if r_squared >= ideal_r:
            sns.lmplot(x = 'Total_Aid_log', y = outcome, data = filtered_data)
            plt.xlabel('Total Aid (log)')
            plt.ylabel(outcome)
            plt.title(f'Total Aid vs. {outcome} in {country} (R-Squared: {r_squared:.2f})')
            plt.show()
```

From the several graphs above, we see a strong positive R-squared correlation in the factors we measured (GDP, GDP per capita, life expectancy) particularly in 2 countries, the Congo Democratic Republic and Nigeria, which are African countries. Out of the 10 countries, this region particularly seems to be positively correlated and responsive to the amount of foreign aid it receives.

An anomaly from this graph is the **unemployment** metric where it shows that Ethiopia has a high R-squared value, suggesting a strong positive correlation between foreign aid received and unemployment numbers. However, this statistic is not significant enough due to the fact that there is not enough data points in the unemployment .csv file for Ethiopia.

### Examining the relationship between indicators and total aid on a country-by-country basis

```python
# Ols model to determine relationship between indicators and total aid by country
# Plot of R-Squared values for each country by indicator 

outcomes = ['GDP_log', 'GDP_Per_Cap_log', 'GDP_Growth_log', 'GDP_Growth_Per_Cap_log',
              'Co2_Emissions_log', 'Life_Expec_log', 'Net_Migration_log', 'Unemployment_log']

for outcome in outcomes:
    r_squared = []
    
    for country, group_country in clean_data.groupby('Country Name'):
        formula = outcome +' ~ Total_Aid_log'
        model = smf.ols(formula, data = group_country).fit()
        r_square = model.rsquared
        r_squared.append(r_square)
        
    plt.figure(figsize = (10,5))
    plt.bar(range(len(r_squared)), r_squared)
    plt.xlabel('Country')
    plt.ylabel('R-Squared')
    plt.title('R-squared for Indicator: ' + outcome)
    plt.xticks(range(len(r_squared)), clean_data['Country Name'].unique(), rotation = 70)
    plt.show()
```

From the bar charts above, we examined the r-squared values.

- The above holds for **GDP per capita**, with most of the countries having about a 0.4 or greater r-squared value (excluding Iraq and Israel), suggesting that foreign aid increases GDP per capita.
- For **GDP growth**, however there was a bit more variety, with the highest r-squared values being about 0.35 in Syria, Nigeria, and Jordan. The rest averaged at an r-squared value of about 0.10, meaning there was weak positive correlation between GDP growth and total foreign aid. In Afghanistan, the Congo Dem. Repub., and Egypt, there was almost no correlation. However, there are other factors that we should consider, like external economic, geographic, or political factors like war that we will examine below.
- **GDP growth per capita** seemed to have moderately low correlation as well, with Syria and South Africa benefitting the most in this metric with the amount of foreign aid received. Egypt, the Congo Dem. Repub., Ethiopia, and Jordan had little to no correlation.
- **CO2 Emissions**, for most countries, were highly positively correlated with total foreign aid received, with most countries excluding Iraq, Israel, and Nigeria having a r-squared value of >0.40, suggesting that in these countries, more foreign aid meant higher CO2 emissions.
- For **Life Expectancy**, most countries had a positive r-squared value of >0.30, excluding Iraq, suggesting that more foreign aid correlates to high life expectancy. Since Iraq is the only anomaly in this group, we will examine external factors (like war) that might answer this later on in our analysis.
- **Net Migration**, for the countries examined excluding South Africa and Afghanistan, seemed to have little to no correlation with total foreign aid received, with said countries having an r-squared value of <0.10.

# Polynomial Regression

To further investigate the any anomalies and any non-linear relationships between variables, we fitted a polynomial regression line into the overall data. The chart below gives the regression line.

```python
# Polynomial Regression, exploring relationship between Total Aid and indicators 

x = clean_data_imputed['Total_Aid_log']
y = clean_data_imputed[['GDP_log', 'GDP_Per_Cap_log', 'GDP_Growth_log', 'GDP_Growth_Per_Cap_log',
                'Co2_Emissions_log', 'Life_Expec_log', 'Net_Migration_log', 'Unemployment_log']]

models = []
degrees = 3 #3 looked best

for column in y.columns:
    coefficients = np.polyfit(x, y[column], degrees)
    model = np.poly1d(coefficients)
    models.append(model)
    r_squared = r2_score(y[column], model(x))
    print("R-squared for", column, ":", r_squared)
    
    plt.scatter(x, y[column])
    plt.plot(x, model(x))
    plt.xlabel('Total Aid (log)')
    plt.ylabel(column)
    plt.title(f'Relationship between Total Aid and {column}')
    plt.show()
```

Based on the results of our polynomial regression, *there was little to no correlation (very low r-squared values) between foreign aid and all metrics (except life expectancy)* across all countries overall. 

It is important to identify life expectancy as positively correlating to total foreign aid received, with a 0.202 r-squared value. This suggests that there could be a LOW positive, linear relationship between life expectancy and total foreign aid received, consistent in all countries according to the data we worked with.
